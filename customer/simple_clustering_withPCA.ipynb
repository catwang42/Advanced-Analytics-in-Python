```python
import pandas as pd
```

```python
df_order = pd.read_excel('WineKMC.xlsx')
```

```python
df_order.columns = ['offer_id','campagin','varietal','min_qty','discount','origin','past_peak']
df_order.head(5)
```

```python
df_trans = pd.read_excel('WineKMC.xlsx',sheetname=1)
df_trans.columns = ['customer_name','offer_id']
df_trans['n'] = 1
df_trans.head(5)
```

```python
df = pd.merge(df_order,df_trans)
df.head(5)
```

```python
#pivot table 
matrix = df.pivot_table(index=['customer_name'],columns = ['offer_id'], values='n')
matrix = matrix.fillna(0).reset_index()
matrix
```

```python
x_cols = matrix.columns[1:]
```

```python
x_cols
```

```python
matrix.columns[2:]
```

```python
from sklearn.cluster import KMeans
culster = KMeans(n_clusters = 5)
matrix['cluster'] = culster.fit_predict(matrix[matrix.columns[1:]])
matrix.cluster.value_counts()
```

```python
from sklearn.decomposition import PCA 

pca = PCA(n_components=2)
matrix['x'] = pca.fit_transform(matrix[x_cols])[:,0]
matrix['y'] = pca.fit_transform(matrix[x_cols])[:,1]
matrix = matrix.reset_index()

customer_clusters = matrix[['customer_name','cluster','x','y']]
customer_clusters.head()
```

```python
df = pd.merge(customer_clusters,df)
```

```python
clour = df['cluster'].values
type(clour)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
x = df['x'].values
y = df['y'].values
clour = df['cluster'].values
#plt.scatter(x, y, c=t)

plt.scatter(x,y,c=clour)
plt.title("Customers Grouped by Cluster")
```

```python
cluster_center = pca.transform(culster.cluster_centers_)
cluster_center = pd.DataFrame(cluster_center,columns=['x', 'y'])
cluster_center['cluster'] = range(0, len(cluster_center))
cluster_center
```

```python
x = df['x'].values
y = df['y'].values
clour = df['cluster'].values
#plt.scatter(x, y, c=t)
x_c = cluster_center['x'].values
y_c = cluster_center['y'].values
clour_c = cluster_center['cluster'].values
plt.scatter(x,y,c=clour)
plt.scatter(x_c,y_c,c=clour_c,s=150,marker='x')
plt.title("Customers Grouped by Cluster")
```
